```python
from tkinter import * 
def Department_Name(vara):
    Department_Frame = Tk()
    
    Department_Frame.geometry("2000x2000")
    i = 0
    Name = Label(Department_Frame,text = "Organizational Structure",bg = "Black",fg = "Red",font = "Times 50",width = 100)
    Name.pack()
    if (vara == "One"):
        Looping_vara = 2
    if (vara == "Two"):
        Looping_vara = 3
    if (vara == "Three"):
        Looping_vara = 4
    if (vara == "Four"):
        Looping_vara = 5
    if (vara == "Five"):
        Looping_vara = 6
        
    listing = ["Head of Organization","1st Department Name","2nd Department Name","3rd Department Name","4th Department Name","5th Department Name"]
    
    while( i < Looping_vara):
        Label_1 = Label (Department_Frame,text =listing[i],font = "Times 30 italic bold")
        entry_Name = Entry(Department_Frame,width = 100)
        Label_1.pack()
        entry_Name.pack(ipady=5)
        Head_and_Department_Name_list.append(entry_Name)
        i = i + 1
        
    button1 = Button(Department_Frame,text= "Submit",fg = "Red",font = "Arial 15 bold",width = 20,height =3,bg = "Black",command =Sub_Divisional_Member)
    button1.pack(padx =5,pady = 10) 
    button2 = Button(Department_Frame,text= "Quit",fg = "Red",font = "Arial 15 bold",width = 20,height =3,bg = "Black",command = Department_Frame.destroy)
    #inserting on the Stack    
    button2.pack()
    
    
    Department_Frame.mainloop()
```

```python
from tkinter import * 
def Submit_button():
# This Function is called When the Submit Button is Pressed  
    toplevel= Tk()
    
    toplevel.geometry("2000x2000")
    topFrame = Frame(toplevel)
    Name = Label(toplevel,text = "Select Structure", fg = "Red",font = "Times 50",bg = "black",width = 100)
    Name.pack()
#Following code is to get value from task bar of the Text
    Functional = IntVar()
    Checkbutton(toplevel,text = "Functional Structure",variable = Functional,font = "Times 30 italic bold",command = new_function).pack()
    Product = IntVar()
    Checkbutton(toplevel,text = "Product Structure",variable = Product,font = "Times 30 italic bold").pack()
    Divisonal = IntVar()
    Checkbutton(toplevel,text = "Divisional Structure",variable = Divisonal,font = "Times 30 italic bold").pack()
    Market = IntVar()
    Checkbutton(toplevel,text = "Market Structure",variable = Market,font = "Times 30 italic bold").pack()
    Matrix = IntVar()
    Checkbutton(toplevel,text = "Divisional Structure",variable = Matrix,font = "Times 30 italic bold").pack()

    button1 = Button(toplevel,text= "Back",fg = "Red",font = "Arial 15 bold",width = 10,height = 2,bg = "Black",command = Main_function)
    button1.pack(padx =5,pady = 10) 
    button2 = Button(toplevel,text= "Quit",fg = "Red",font = "Arial 15 bold",width = 10,height = 2,bg = "Black",command = toplevel.destroy)
    #inserting on the Stack    
    button2.pack()
    toplevel.mainloop()
    return

```
